```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import concurrent.futures
import pandas as pd
```

```python
url = 'https://www.macys.com/shop/womens-clothing/all-womens-shoes/flats?id=50295'

ua = UserAgent()
user_agent = ua.random
headers = {'User-Agent': user_agent}

page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.content, 'html.parser')

# print("Status Code:", page.status_code)


sales_box = soup.find_all('div', class_ = 'product-thumbnail-container vertical-alignment')
```

```python
product_brands = []
product_names = []
discount_prices = []
discount_prices_with_promo = []
old_prices = []
ratings = []
product_urls = []

#product_brands
for box in sales_box:
    if box.find('div', class_ = 'product-brand medium') is not None:
        product_brand = box.find('div', class_ = 'product-brand medium')
        product_brands.append(product_brand.text)
    else:
        product_brands.append('None')

#product_names
    if box.find('div', class_ = 'product-name medium') is not None:
        product_name = box.find('div', class_ = 'product-name medium')
        product_names.append(product_name.text.lstrip())
    else:
        product_names.append('None')

#discount_prices
    if box.find('span', class_ = 'discount is-tier2') is not None:
        discount_price = box.find('span', class_ = 'discount is-tier2')
        discount_prices.append(discount_price.text)
    else:
        discount_prices.append('None')

#discount_prices_with_promo_code
    if box.find('div', class_ = 'discount price-reg badges-pricing-ind-tier') is not None:
        discount_price_with_promo = box.find('div', class_ = 'discount price-reg badges-pricing-ind-tier')
        discount_prices_with_promo.append(discount_price.text)
    else:
        discount_prices_with_promo.append('None')

#Old prices
    if box.find('span', class_ = 'price-strike-sm') is not None:
        old_price = box.find('span', class_ = 'price-strike-sm')
        old_prices.append(old_price.text.strip())
    else:
        old_prices.append('None')

#rating
    if box.select_one('div.rating > span.review-star-wrapper') is not None:
        rating = box.select_one('div.rating > span.review-star-wrapper')
        rating = rating['aria-label']
        ratings.append(rating.replace('Rated', '').replace(' stars','').strip())
    else:
        ratings.append('None')

#product_url
    if box.find('a', class_ = 'brand-and-name') is not None:
        product_url = box.find('a', class_ = 'brand-and-name')
        product_url = product_url['href']
        product_urls.append('https://www.macys.com' + product_url)
    else:
        product_urls.append('None')


product_urls
```

```python
data = pd.DataFrame({
    'Product Brand': product_brands,
    'Product Name': product_names,
    'Discount_Prices': discount_prices,
    'Discount Prices with Promo Code': discount_prices_with_promo,
    'Old Price': old_prices,
    'Rating': ratings,
    'Product URL': product_urls
})
data
```
